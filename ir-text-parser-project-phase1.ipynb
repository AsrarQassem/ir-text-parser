## 1- Tokenizer: Reads document into memory, tokenizes to separate words; returns token stream. Basic tokenization rules:
- Remove numbers
- Ignore if word contains numbers. 
- Split on all nonalphanumeric characters(such as punctuation marks, spaces,
   hyphens, and apostrophes)
- Convert to lower case

After reading Document content by `process_file()`, the following tokenizer is going to apply all previous rules

````python
#Tonkenizer:
def tokenize(doc_content):
    # Remove the DOCNO and PROFILE tags to avoid including it in the tokens
    doc_content = re.sub(r'<DOCNO>.*?</DOCNO>', '', doc_content, flags=re.DOTALL)
    doc_content = re.sub(r'<PROFILE>.*?</PROFILE>', '', doc_content, flags=re.DOTALL)
    tokens = re.split(r'\W+', doc_content)
    tokens = [token.lower() for token in tokens if token.isalpha()]
    return tokens
````

## Stopwords:
this function is going to read the stopwords file **stopwordlist.txt** and return them as set.

````python
# load stopwords:
def load_stopwords(file_path):
    with open(file_path, 'r') as f:
        return set(line.strip() for line in f.read().splitlines())
````

## Document Preprocessing (Data handling):
The parser processes multiple files from a directory, handling each file's content individually to ensure proper tokenization and indexing of documents. This function reads each TREC file and uses regular expression **re** to find the Doc number in the tags `<DOCNO>` to be used as Doc name. then it reads the entire text in each `<DOC>` and pass entire content to the tokenizer to be return as tokens after applying tokenizer rules.

````python
# To read files and bring content of file into memory:
def process_file(file_path, file_dict, word_dict):
    with open(file_path, 'r', encoding='utf-8', errors='ignore') as f:
        file_content = f.read()
    
    #for data -> parse them to separate documents.
    # get Doc and Doc ID
    documents = re.findall(r'<DOC>(.*?)</DOC>', file_content, re.DOTALL)
    for doc in documents:
        doc_no = re.search(r'<DOCNO>(.*?)</DOCNO>', doc, re.DOTALL).group(1).strip()

        # To map each document name to a unique numerical ID. 
        file_dict[doc_no] = len(file_dict) + 1
        # Tokenizing entire content of <DOC>
        tokens = tokenize(doc)
        for token in tokens:

            #To eliminate stopwords
            if token not in stop_words:
                # Stemming of the words 
                stemmed_token = stemmer.stem(token)
                #Selection of terms for the term dictionary
                if stemmed_token not in word_dict:
                    #map each token or term to a unique numerical ID.
                    #Creating the dictionary file (Term Dictionary and Document Dictionary)
                    word_dict[stemmed_token] = len(word_dict) + 1
````
